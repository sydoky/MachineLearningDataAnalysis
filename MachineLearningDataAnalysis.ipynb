```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
housing = pd.read_csv("/content/Housing.csv")
PopChange = pd.read_csv("/content/PopChange.csv")
```

```python
#Build a userinterface



while True:
  print("Welcome to the Python Data Analysis App")
  print("Select the file you want to analyze: ")
  print("1. Population Data")
  print("2. Housing Data")
  print("3. Exit the Program")
  choose_option=(input("Select your option"))

  if choose_option == "1":
  
    while True:
      print("You have entered Population Data.")
      print("Select the Column you want to analyze:")
      print("a. Pop Apr 1")
      print("b. Pop Jul 1")
      print("c. Change Pop")
      print("d. Exit Column")
      choose_option2=(input("Select the column"))
      if choose_option2=="a":
        print("You selected Pop Apr 1")
        print("The statistics for this column are: ")
        print("count = {}".format(PopChange["Pop Apr 1"].count()))
        print("mean = {}".format(PopChange["Pop Apr 1"].mean()))
        print("standard deviation = {}".format(PopChange["Pop Apr 1"].std()))
        print("min = {}".format(PopChange["Pop Apr 1"].min()))
        print("max = {}".format(PopChange["Pop Apr 1"].max()))
        print("The Histogram of this column is now displayed")
        PopChange["Pop Apr 1"].hist()
        plt.show()
      elif choose_option2=="b":
        print("You selected Pop Jul 1")
        print("The statistics for this column are: ")
        print("count = {}".format(PopChange["Pop Jul 1"].count()))
        print("mean = {}".format(PopChange["Pop Jul 1"].mean()))
        print("standard deviation = {}".format(PopChange["Pop Jul 1"].std()))
        print("min = {}".format(PopChange["Pop Jul 1"].min()))
        print("max = {}".format(PopChange["Pop Jul 1"].max()))
        print("The Histogram of this column is now displayed")
        PopChange["Pop Jul 1"].hist()
        plt.show()
      elif choose_option2=="c":
        print("You selected Change Pop")
        print("The statistics for this column are: ")
        print("count = {}".format(PopChange["Change Pop"].count()))
        print("mean = {}".format(PopChange["Change Pop"].mean()))
        print("standard deviation = {}".format(PopChange["Change Pop"].std()))
        print("min = {}".format(PopChange["Change Pop"].min()))
        print("max = {}".format(PopChange["Change Pop"].max()))
        print("The Histogram of this column is now displayed")
        PopChange["Change Pop"].hist()
        plt.show()
      elif choose_option2=="d":
        print("You selected Exit Column")
        break
      else:
        print("Invalid Input")

  elif choose_option == "2":

    while True:
      print("You have entered Housing Data.")
      print("Select the Column you want to analyze:")
      print("a. AGE")
      print("b.BEDRMS")
      print("c.BUILT")
      print("d.ROOMS")
      print("e.UTILITY")
      print("f.Exit the program")
      choose_option2=(input("Select the column"))
      if choose_option2=="a":
        print("You selected AGE")
        print("The statistics for this column are: ")
        print("count = {}".format(housing["AGE"].count()))
        print("mean = {}".format(housing["AGE"].mean()))
        print("standard deviation = {}".format(housing["AGE"].std()))
        print("min = {}".format(housing["AGE"].min()))
        print("max = {}".format(housing["AGE"].max()))
        print("The Histogram of this column is now displayed")
        housing["AGE"].hist()
        plt.show()
      if choose_option2=="b":
        print("You selected BEDRMS")
        print("The statistics for this column are: ")
        print("count = {}".format(housing["BEDRMS"].count()))
        print("mean = {}".format(housing["BEDRMS"].mean()))
        print("standard deviation = {}".format(housing["BEDRMS"].std()))
        print("min = {}".format(housing["BEDRMS"].min()))
        print("max = {}".format(housing["BEDRMS"].max()))
        print("The Histogram of this column is now displayed")
        housing["BEDRMS"].hist()
        plt.show()
      if choose_option2=="c":
        print("You selected BUILT")
        print("The statistics for this column are: ")
        print("count = {}".format(housing["BUILT"].count()))
        print("mean = {}".format(housing["BUILT"].mean()))
        print("standard deviation = {}".format(housing["BUILT"].std()))
        print("min = {}".format(housing["BUILT"].min()))
        print("max = {}".format(housing["BUILT"].max()))
        print("The Histogram of this column is now displayed")
        housing["BUILT"].hist()
        plt.show()
      if choose_option2=="d":
        print("You selected ROOMS")
        print("The statistics for this column are: ")
        print("count = {}".format(housing["ROOMS"].count()))
        print("mean = {}".format(housing["ROOMS"].mean()))
        print("standard deviation = {}".format(housing["ROOMS"].std()))
        print("min = {}".format(housing["ROOMS"].min()))
        print("max = {}".format(housing["ROOMS"].max()))
        print("The Histogram of this column is now displayed")
        housing["ROOMS"].hist()
        plt.show()
      if choose_option2=="e":
        print("You selected UTILITY")
        print("The statistics for this column are: ")
        print("count = {}".format(housing["UTILITY"].count()))
        print("mean = {}".format(housing["UTILITY"].mean()))
        print("standard deviation = {}".format(housing["UTILITY"].std()))
        print("min = {}".format(housing["UTILITY"].min()))
        print("max = {}".format(housing["UTILITY"].max()))
        print("The Histogram of this column is now displayed")
        housing["UTILITY"].hist()
        plt.show()
      elif choose_option2=="f":
        print("You selected Exit Column")
        break
      else:
        print("Invalid Input")

  elif choose_option == "3":
    print("*************** Thanks for using the Data Analysis App********** ")
    break
  else:
    print("Sorry, Invalid Input")

```
